```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snsperplexit√©
import os
import sys
from sklearn.manifold import TSNE
```

```python
files = [file for file in os.listdir() if file.endswith(".csv") ]
files
```

```python
def plot_tsne(file, perplexity = 25):
  data = pd.read_csv(file, header=None, sep=" ")
  n_col = len(data.columns)

  tsne = TSNE(perplexity=perplexity)
  data_transformed = tsne.fit_transform(data[list(range(n_col))])

  plt.scatter(data_transformed[:, 0], data_transformed[:, 1], c=data.iloc[:, -1])
  plt.title(file)
  plt.show()
```

```python
#perplexit√©=5
for file in files:
  print(file)
  plot_tsne(file)
```

```python
#perplexit√©=10
for file in files:
  print(file)
  plot_tsne(file)
```
